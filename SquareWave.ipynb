```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Generate a square wave signal with white Gaussian noise?

```python
t = np.linspace(2,18,1601,endpoint=True, retstep=False, dtype=None, axis=0)
squareWave = np.sign(np.sin(2*np.pi*0.25*t))   ## frequency is 0.25Hz
squareWave = (squareWave+1)/2

## add white Gaussian noise with variance of 1/10000
squareWave = squareWave + np.multiply(squareWave, np.random.randn(np.size(t))/100)
plt.plot(t,squareWave,'k')
plt.title('Conditional stimulus')
plt.xlabel('time(s)')
plt.ylabel('CS')

plt.subplots_adjust(left = 0.15, right = 0.9)
plt.show()
```
